```python
import sys

# 4 bit
decode = {
    b"0 1 0 1": b"A",
    b"0 1 1 1": b"B",
    b"0 0 1 0": b"C",
    b"1 1 0 1": b"D",
    b"1 0 0 0": b"E",
    b"1 1 0 0": b"F",
}

read = sys.stdin.buffer.readline
write = sys.stdout.buffer.write

while True:
    line = read()

    if not line:
        break

    line = line.strip()

    if not line:
        continue

    n = int(line)

    buf = []

    for _ in range(n):

        code = read().strip()

        buf.append(decode[code])

        if len(buf) >= 8192:
            write(b"".join(buf))
            buf.clear
    
    if buf:
        write(b"".join(buf))
    
    write(b"\n")
```
