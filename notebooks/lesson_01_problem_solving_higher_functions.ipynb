`segments` 是多個線段，每個線段用 `(L, R)` 表示。請回傳所有線段覆蓋的總長度，重疊部分只能算一次。

提示：
1. 先排序。
2. 維護目前合併中的線段左右端點。
3. 遇到沒有重疊的新線段，就把舊線段長度加入答案。

````python
list_1 = [(5,6), (1,2), (4,8), (7,9)]
sorted(list_1)
````

````python
def q9_cover_length(segments):
    # TODO: 回傳總覆蓋長度
    if not segments:
        return 0
    
    segs = sorted(segments)
    total_length = 0
    left, right = segs[0]

    for l,r in segs[1:]:
        if l > right:
            total_length += right  - left
            left, right = l,r
        else:
            if r > right:
                right = r
    
    total_length += right - left


    return total_length
````
