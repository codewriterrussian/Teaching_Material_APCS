## Section II. 基本概念：程式正常跑 vs. 發生錯誤

程式執行時，如果某一行出錯，後面的程式通常就不會繼續執行。

使用 `try / except` 的想法是：

```text
try:
    嘗試執行可能出錯的程式
except 錯誤類型:
    如果真的發生這種錯誤，就改做這裡的處理
```

流程可以想成：

```text
開始
↓
執行 try 裡面的程式
↓
如果沒有錯誤 → 繼續往下
如果發生錯誤 → 跳到 except
↓
程式繼續執行，不會直接當掉
```


## Section III. 範例 1：避免輸入文字造成程式中斷

下面的程式要求使用者輸入整數，然後輸出 `整數 + 1`。

如果使用者輸入的是文字，例如 `hello`，`int(x)` 會發生 `ValueError`。


````python
try:
    x = input("請輸入整數：")
    print(int(x) + 1)
except ValueError:
    print("錯誤！輸入的不是整數！")
````

### 重點說明

- `input()` 讀進來的資料一開始都是字串。
- `int(x)` 嘗試把字串轉成整數。
- 如果無法轉成整數，就會發生 `ValueError`。
- `except ValueError:` 會接住這種錯誤，並執行錯誤處理。


## Section IV. 範例 2：輸入錯誤時，讓使用者重新輸入

使用 `while True` 可以讓程式一直重複執行，直到輸入正確為止。

當輸入成功後，用 `break` 離開迴圈。


````python
while True:
    print("請重新輸入：")
    x = input("請輸入整數：")

    try:
        print(int(x) + 1)
        break
    except ValueError:
        print("錯誤！輸入的不是整數！")
````

### 流程說明

```text
開始 while True
↓
輸入 x
↓
嘗試 int(x)
↓
成功 → 印出結果 → break 離開迴圈
失敗 → 印出錯誤訊息 → 回到迴圈開頭
```


## Section V. 範例 3：最多只給 3 次輸入機會

有時候我們不希望程式無限重複，可以使用 `for` 搭配 `range(3)` 限制最多執行 3 次。


````python
for attempt in range(3):  # 允許最多 3 次輸入
    print("輸入機會剩餘：", 3 - attempt)
    x = input("請輸入整數：")

    try:
        print(int(x) + 1)
        break
    except ValueError:
        print("錯誤！輸入的不是整數！")
````

### 重點說明

- `range(3)` 會產生 `0, 1, 2`，所以迴圈最多執行 3 次。
- `attempt` 表示目前是第幾次嘗試。
- 如果輸入成功，`break` 會提早結束迴圈。


## Section VI. 不建議：直接寫 `except:`

下面的寫法可以接住所有錯誤，但不建議初學者這樣寫，因為它可能會把真正的 bug 也藏起來。


````python
try:
    x = input("請輸入整數：")
    print(int(x) + 1)
except:
    print("錯誤！輸入的不是整數！")
````

### 較好的寫法

知道可能發生哪一種錯誤時，應該明確寫出錯誤類型。

例如：

```python
except ValueError:
```

這樣程式比較清楚，也比較容易除錯。


## Section VII. 範例 4：字典 KeyError

如果字典中沒有某個 key，直接讀取會發生 `KeyError`。


````python
dict_test = {"a": 1, "b": 2}

try:
    print(dict_test["c"])
except KeyError:
    print("錯誤！字典中沒有這個鍵！")
    dict_test["c"] = 3
    print("已添加鍵 'c'，值為 3。")

print(dict_test)
````

### 重點說明

- `dict_test["c"]` 代表讀取 key 是 `"c"` 的值。
- 如果 `"c"` 不存在，就會發生 `KeyError`。
- 可以在 `except KeyError:` 裡面補上預設值。


## Section VIII. 範例 5：處理多種錯誤

同一段程式可能發生不同種類的錯誤。

例如：
- 輸入不是整數 → `ValueError`
- 除以 0 → `ZeroDivisionError`


````python
try:
    a = int(input("請輸入第一個整數："))
    b = int(input("請輸入第二個整數："))
    print(a / b)
except ValueError:
    print("錯誤！請輸入整數！")
except ZeroDivisionError:
    print("錯誤！不能除以 0！")
````

### 重點說明

`except` 可以寫多個，Python 會依照錯誤種類，跳到對應的錯誤處理區塊。

## Section X. 常見錯誤提醒：TypeError vs. ValueError

`input()` 得到的是字串，所以 `int(input_value)` 最常見的錯誤是 `ValueError`。

`TypeError` 通常出現在資料型態不支援某個操作時，例如把不適合的型態拿去運算。


````python
try:
    integer_value = input("請輸入整數：")
    print(int(integer_value) + 1)
except ValueError:
    print("錯誤！輸入的不是整數！")

print(integer_value)
````
