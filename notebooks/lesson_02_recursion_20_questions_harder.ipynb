```python

import base64, json, math

_DATA = """eyJxMV9wcm9kdWN0X3JhbmdlIjogW1sxLCA1XSwgWzMsIDZdLCBbNSwgNV0sIFs3LCAzXV0sICJxMl9jb3VudF9uZXN0ZWQiOiBbW1sxLCBbMiwgMV0sIFszLCBbMSwgNF1dLCAxXSwgMV0sIFtbWyJhIiwgWyJiIiwgImEiXSwgW11dLCAiYSJdLCAiYSJdLCBbW1tdLCBbMiwgWzIsIFsyXV1dXSwgMl0sIFtbMSwgMiwgM10sIDRdXSwgInEzX2ZsYXR0ZW4iOiBbW1sxLCBbMiwgM10sIFtbNF0sIDVdXV0sIFtbW10sIFsxLCBbMl1dLCAzXV0sIFtbXV0sIFtbW1siYSJdXSwgWyJiIiwgWyJjIl1dXV1dLCAicTRfcmVjdXJzaXZlX2xlbiI6IFtbInJlY3Vyc2lvbiJdLCBbIiJdLCBbIlB5dGhvbiDpgZ7ov7QiXSwgWyJhYmMiXV0sICJxNV9yZW1vdmVfY2hhciI6IFtbImJhbmFuYSIsICJhIl0sIFsicmVjdXJzaW9uIiwgInIiXSwgWyJhYWFhIiwgImEiXSwgWyJoZWxsbyIsICJ4Il1dLCAicTZfaXNfc29ydGVkIjogW1tbMSwgMiwgMywgMywgNV1dLCBbWzFdXSwgW1tdXSwgW1szLCAyLCAxXV0sIFtbMSwgMywgMl1dXSwgInE3X2JpbmFyeV9zZWFyY2giOiBbW1sxLCAzLCA1LCA3LCA5XSwgNywgMCwgNF0sIFtbMSwgMywgNSwgNywgOV0sIDIsIDAsIDRdLCBbWzEwXSwgMTAsIDAsIDBdLCBbWzEwXSwgNSwgMCwgMF0sIFtbMiwgNCwgNiwgOCwgMTAsIDEyXSwgMTIsIDAsIDVdXSwgInE4X21lcmdlX3NvcnQiOiBbW1szLCAxLCA0LCAxLCA1XV0sIFtbXV0sIFtbMV1dLCBbWzksIC0xLCAwLCA5LCAzXV1dLCAicTlfZmFzdF9wb3dlcl9tb2QiOiBbWzIsIDEwLCAxMDAwXSwgWzMsIDAsIDddLCBbNSwgMywgMTNdLCBbMTAsIDgsIDE3XSwgWzcsIDExLCAxOV1dLCAicTEwX2djZF9saXN0IjogW1tbMTIsIDE4LCAzMF1dLCBbWzddXSwgW1s4MSwgMjcsIDldXSwgW1sxNCwgMzUsIDQ5XV1dLCAicTExX2RpZ2l0YWxfcm9vdCI6IFtbMF0sIFs5XSwgWzk4NzVdLCBbMTIzNDU2Nzg5XSwgWzk5OTk5OV1dLCAicTEyX2RlY2ltYWxfdG9fYmFzZSI6IFtbMCwgMl0sIFsxMCwgMl0sIFszMSwgMTZdLCBbMjU1LCAxNl0sIFsxMDAsIDhdLCBbMzUsIDM2XV0sICJxMTNfY291bnRfcGF0aHMiOiBbWzEsIDFdLCBbMSwgNV0sIFszLCAzXSwgWzMsIDRdLCBbNCwgNV1dLCAicTE0X2hhbm9pIjogW1sxLCAiQSIsICJCIiwgIkMiXSwgWzIsICJBIiwgIkIiLCAiQyJdLCBbMywgIkwiLCAiTSIsICJSIl1dLCAicTE1X2ZpYl9tZW1vIjogW1swXSwgWzFdLCBbMTBdLCBbMjBdLCBbMzBdXSwgInExNl90cmlib25hY2NpIjogW1swXSwgWzFdLCBbMl0sIFs1XSwgWzhdLCBbMTBdXSwgInExN19wYXNjYWxfcm93IjogW1swXSwgWzFdLCBbNF0sIFs2XV0sICJxMThfYWxsX3N1YnNldHMiOiBbW1tdXSwgW1sxXV0sIFtbMSwgMl1dLCBbWzEsIDIsIDNdXV0sICJxMTlfcGVybXV0YXRpb25zIjogW1siYSJdLCBbImFiIl0sIFsiYWJjIl0sIFsiYWFiIl1dLCAicTIwX2dlbmVyYXRlX3BhcmVudGhlc2VzIjogW1swXSwgWzFdLCBbMl0sIFszXV19"""
_CASES = {k: [tuple(x) for x in v] for k, v in json.loads(base64.b64decode(_DATA).decode()).items()}


def _report(name, ok, detail=""):
    if ok:
        print(f"✅ {name} pass")
    else:
        print(f"❌ {name} fail")
        if detail:
            print(detail)


def _ref_product_range(a, b):
    if a > b:
        return 1
    return a * _ref_product_range(a + 1, b)


def _ref_count_nested(data, target):
    if data == []:
        return 0
    first, rest = data[0], data[1:]
    if isinstance(first, list):
        return _ref_count_nested(first, target) + _ref_count_nested(rest, target)
    return (1 if first == target else 0) + _ref_count_nested(rest, target)


def _ref_flatten(data):
    if data == []:
        return []
    first, rest = data[0], data[1:]
    if isinstance(first, list):
        return _ref_flatten(first) + _ref_flatten(rest)
    return [first] + _ref_flatten(rest)


def _ref_recursive_len(s):
    if s == "":
        return 0
    return 1 + _ref_recursive_len(s[1:])


def _ref_remove_char(s, ch):
    if s == "":
        return ""
    head = "" if s[0] == ch else s[0]
    return head + _ref_remove_char(s[1:], ch)


def _ref_is_sorted(nums):
    if len(nums) <= 1:
        return True
    if nums[0] > nums[1]:
        return False
    return _ref_is_sorted(nums[1:])


def _ref_binary_search(nums, target, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if nums[mid] == target:
        return mid
    if target < nums[mid]:
        return _ref_binary_search(nums, target, left, mid - 1)
    return _ref_binary_search(nums, target, mid + 1, right)


def _merge(a, b):
    if not a:
        return b[:]
    if not b:
        return a[:]
    if a[0] <= b[0]:
        return [a[0]] + _merge(a[1:], b)
    return [b[0]] + _merge(a, b[1:])


def _ref_merge_sort(nums):
    if len(nums) <= 1:
        return nums[:]
    mid = len(nums) // 2
    return _merge(_ref_merge_sort(nums[:mid]), _ref_merge_sort(nums[mid:]))


def _ref_fast_power_mod(x, n, mod):
    if n == 0:
        return 1 % mod
    half = _ref_fast_power_mod(x, n // 2, mod)
    ans = (half * half) % mod
    if n % 2 == 1:
        ans = (ans * x) % mod
    return ans


def _gcd(a, b):
    return abs(a) if b == 0 else _gcd(b, a % b)


def _ref_gcd_list(nums):
    if len(nums) == 1:
        return abs(nums[0])
    return _gcd(nums[0], _ref_gcd_list(nums[1:]))


def _sum_digits(n):
    if n < 10:
        return n
    return n % 10 + _sum_digits(n // 10)


def _ref_digital_root(n):
    if n < 10:
        return n
    return _ref_digital_root(_sum_digits(n))


def _ref_decimal_to_base(n, base):
    chars = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if n < base:
        return chars[n]
    return _ref_decimal_to_base(n // base, base) + chars[n % base]


def _ref_count_paths(m, n):
    if m == 1 or n == 1:
        return 1
    return _ref_count_paths(m - 1, n) + _ref_count_paths(m, n - 1)


def _ref_hanoi(n, src, aux, dst):
    if n == 0:
        return []
    return _ref_hanoi(n - 1, src, dst, aux) + [(src, dst)] + _ref_hanoi(n - 1, aux, src, dst)


def _ref_fib_memo(n, memo=None):
    if memo is None:
        memo = {}
    if n in memo:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = _ref_fib_memo(n - 1, memo) + _ref_fib_memo(n - 2, memo)
    return memo[n]


def _ref_tribonacci(n):
    if n == 0:
        return 0
    if n == 1 or n == 2:
        return 1
    return _ref_tribonacci(n - 1) + _ref_tribonacci(n - 2) + _ref_tribonacci(n - 3)


def _ref_pascal_row(n):
    if n == 0:
        return [1]
    prev = _ref_pascal_row(n - 1)
    return [1] + [prev[i] + prev[i + 1] for i in range(len(prev) - 1)] + [1]


def _ref_all_subsets(nums):
    if len(nums) == 0:
        return [[]]
    rest = _ref_all_subsets(nums[1:])
    return rest + [[nums[0]] + subset for subset in rest]


def _ref_permutations(s):
    if len(s) <= 1:
        return [s]
    ans = []
    for i, ch in enumerate(s):
        rest = s[:i] + s[i+1:]
        for p in _ref_permutations(rest):
            ans.append(ch + p)
    return ans


def _ref_generate_parentheses(n):
    def helper(open_left, close_left):
        if open_left == 0 and close_left == 0:
            return [""]
        ans = []
        if open_left > 0:
            for s in helper(open_left - 1, close_left):
                ans.append("(" + s)
        if close_left > open_left:
            for s in helper(open_left, close_left - 1):
                ans.append(")" + s)
        return ans
    return helper(n, n)

_REF = {
    'q1_product_range': _ref_product_range,
    'q2_count_nested': _ref_count_nested,
    'q3_flatten': _ref_flatten,
    'q4_recursive_len': _ref_recursive_len,
    'q5_remove_char': _ref_remove_char,
    'q6_is_sorted': _ref_is_sorted,
    'q7_binary_search': _ref_binary_search,
    'q8_merge_sort': _ref_merge_sort,
    'q9_fast_power_mod': _ref_fast_power_mod,
    'q10_gcd_list': _ref_gcd_list,
    'q11_digital_root': _ref_digital_root,
    'q12_decimal_to_base': _ref_decimal_to_base,
    'q13_count_paths': _ref_count_paths,
    'q14_hanoi': _ref_hanoi,
    'q15_fib_memo': _ref_fib_memo,
    'q16_tribonacci': _ref_tribonacci,
    'q17_pascal_row': _ref_pascal_row,
    'q18_all_subsets': _ref_all_subsets,
    'q19_permutations': _ref_permutations,
    'q20_generate_parentheses': _ref_generate_parentheses,
}

_UNORDERED_LIST_OUTPUTS = {'q18_all_subsets', 'q19_permutations', 'q20_generate_parentheses'}


def _normalize(name, value):
    if name == 'q18_all_subsets':
        return sorted([tuple(x) for x in value])
    if name in {'q19_permutations', 'q20_generate_parentheses'}:
        return sorted(value)
    return value


def _check_q(name):
    if name not in globals():
        _report(name, False, f"找不到函式 {name}，請確認函式名稱沒有改掉。")
        return
    if name not in _REF:
        _report(name, False, f"檢查工具沒有這一題：{name}")
        return
    func = globals()[name]
    ref = _REF[name]
    for args in _CASES[name]:
        try:
            got = func(*args)
        except RecursionError:
            _report(name, False, f"測資 {args} 發生 RecursionError，可能終止條件或問題變小有錯。")
            return
        except Exception as e:
            _report(name, False, f"測資 {args} 發生錯誤：{type(e).__name__}: {e}")
            return
        expected = ref(*args)
        if _normalize(name, got) != _normalize(name, expected):
            _report(name, False, f"測資 {args}\n你的輸出：{got}\n預期輸出：{expected}")
            return
    _report(name, True)
```
