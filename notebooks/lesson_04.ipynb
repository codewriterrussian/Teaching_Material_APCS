## Section II. 什麼是遞迴？

遞迴就是：**函式在自己的程式裡面，再呼叫自己一次。**

例如：

```python
def f(n):
    return f(n - 1)
```

但是這樣會有一個問題：

> 如果一直呼叫自己，什麼時候才會停？

所以遞迴一定要有「停下來的條件」。

## Section III. 遞迴的兩個重要部分

寫遞迴時，可以先問自己兩個問題：

### 1. 什麼時候不用再拆？

這叫做 **base case（終止條件）**。

例如階乘：

```text
1! = 1
```

所以 `factorial(1)` 可以直接回傳 `1`。

### 2. 其他情況怎麼變成更小的問題？

這叫做 **recursive case（遞迴關係）**。

例如：

```text
5! = 5 × 4!
4! = 4 × 3!
3! = 3 × 2!
2! = 2 × 1!
```

也就是：

```text
n! = n × (n - 1)!
```

````python
def factorial(n):
    # base case：最小問題，直接回答
    if n == 1:
        return 1

    # recursive case：把問題變小
    return n * factorial(n - 1)

print(factorial(5))
````

## Section IV. `factorial(5)` 是怎麼算的？

程式會先一路往下拆：

```text
factorial(5)
= 5 × factorial(4)
= 5 × 4 × factorial(3)
= 5 × 4 × 3 × factorial(2)
= 5 × 4 × 3 × 2 × factorial(1)
```

遇到 base case：

```text
factorial(1) = 1
```

然後再把答案一層一層回傳回去：

```text
factorial(2) = 2 × 1 = 2
factorial(3) = 3 × 2 = 6
factorial(4) = 4 × 6 = 24
factorial(5) = 5 × 24 = 120
```

可以把遞迴想成：

> 先一直問更小的問題，直到最小問題有答案，再把答案一路傳回來。

````python
def factorial_trace(n, depth=0):
    indent = "  " * depth
    print(indent + f"進入 factorial({n})")

    if n == 1:
        print(indent + "遇到 base case，回傳 1")
        return 1

    result = n * factorial_trace(n - 1, depth + 1)
    print(indent + f"factorial({n}) 回傳 {result}")
    return result

factorial_trace(5)
````

## Section V. 小心：沒有正確 base case 會爆掉

下面這種寫法很危險：

```python
def bad_factorial(n):
    return n * bad_factorial(n - 1)
```

因為它永遠不會停：

```text
bad_factorial(5)
bad_factorial(4)
bad_factorial(3)
bad_factorial(2)
bad_factorial(1)
bad_factorial(0)
bad_factorial(-1)
...
```

最後會出現 `RecursionError`。

## Section VI. 更完整的階乘寫法

一般來說，`0! = 1`，所以我們可以讓 `n == 0` 或 `n == 1` 都回傳 `1`。

````python
def factorial(n):
    if n < 0:
        return "階乘不接受負數"

    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)

for i in range(6):
    print(i, factorial(i))
````

## Section VII. 組合數 `C(n, m)`

組合數常寫成：

```text
C(n, m)
```

意思是：

> 從 n 個東西裡面，選 m 個出來，有幾種選法？

例如：

```text
C(5, 2)
```

代表從 5 個人裡面選 2 個人。

---

## Section VIII. 組合數的公式法

公式是：

```text
C(n, m) = n! / (m! × (n - m)!)
```

例如：

```text
C(5, 2) = 5! / (2! × 3!) = 10
```

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


def C_formula(n, m):
    numerator = factorial(n)
    denominator = factorial(m) * factorial(n - m)
    return numerator // denominator

print(C_formula(5, 2))
print(C_formula(6, 3))
````

## Section IX. 組合數的遞迴想法

除了公式法，組合數也可以用遞迴理解。

假設我們要算：

```text
C(n, m)
```

可以想成：第 n 個人有兩種情況。

### 情況 1：選第 n 個人

如果選了第 n 個人，剩下要從前 `n - 1` 個人裡面選 `m - 1` 個。

```text
C(n - 1, m - 1)
```

### 情況 2：不選第 n 個人

如果不選第 n 個人，就要從前 `n - 1` 個人裡面選 `m` 個。

```text
C(n - 1, m)
```

所以：

```text
C(n, m) = C(n - 1, m - 1) + C(n - 1, m)
```

---

## Section X. 組合數的 base case

什麼時候可以直接回答？

```text
C(n, 0) = 1
```

因為從 n 個東西裡面選 0 個，只有 1 種選法：什麼都不選。

```text
C(n, n) = 1
```

因為從 n 個東西裡面選 n 個，也只有 1 種選法：全部都選。

````python
def C(n, m):
    # 不合法情況
    if m < 0 or m > n:
        return 0

    # base case
    if m == 0 or m == n:
        return 1

    # recursive case
    return C(n - 1, m - 1) + C(n - 1, m)

print(C(5, 2))
print(C(6, 3))
````

## Section XI. 用遞迴畫出 Pascal Triangle

Pascal Triangle（巴斯卡三角形）的每個數字都可以用組合數表示。

例如第 5 列：

```text
C(5,0) C(5,1) C(5,2) C(5,3) C(5,4) C(5,5)
1      5      10     10     5      1
```

````python
def C(n, m):
    if m < 0 or m > n:
        return 0
    if m == 0 or m == n:
        return 1
    return C(n - 1, m - 1) + C(n - 1, m)

for n in range(7):
    row = []
    for m in range(n + 1):
        row.append(C(n, m))
    print(row)
````

## Section XII. 費氏數列 Fibonacci

費氏數列的規則是：

```text
fib(0) = 0
fib(1) = 1
fib(n) = fib(n - 1) + fib(n - 2)
```

前幾項是：

```text
0, 1, 1, 2, 3, 5, 8, 13, ...
```

這也很適合用遞迴表示，因為 `fib(n)` 需要前面兩個更小的答案。

````python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

for i in range(10):
    print(f"fib({i}) = {fib(i)}")
````

## Section XIII. Fibonacci 的問題：重複計算很多次

遞迴版 Fibonacci 很好懂，但是效率不好。

例如 `fib(5)` 會拆成：

```text
fib(5)
= fib(4) + fib(3)
= fib(3) + fib(2) + fib(2) + fib(1)
```

你會發現 `fib(3)`、`fib(2)` 可能被重複算很多次。

所以 Fibonacci 適合用來學遞迴觀念，但如果要算很大的 `n`，要用更有效率的方法。

````python
def fib_loop(n):
    a = 0
    b = 1

    for _ in range(n):
        a, b = b, a + b

    return a

for i in range(10):
    print(f"fib_loop({i}) = {fib_loop(i)}")
````

## Section XIV. 把文字規則轉成程式

原始規則：

```text
若 x = 1，則 F(x) = 1
若 x 為偶數，則 F(x) = F(x / 2)
其餘狀況，F(x) = F(x - 1) + F(x + 1)
```

這種題目要先找：

1. base case 是什麼？
2. 哪些情況會呼叫自己？
3. 每次呼叫後，問題有沒有變得更接近 base case？

這題有一個重要問題：

```text
F(x) = F(x - 1) + F(x + 1)
```

如果直接照寫，可能會一直往更大的數字跑，導致遞迴停不下來。

例如 `F(3)`：

```text
F(3) = F(2) + F(4)
F(4) = F(2)
```

這裡剛好可以停，但很多遞迴題都要特別檢查「會不會越跑越遠」。

````python
def F(x):
    if x == 1:
        return 1

    if x % 2 == 0:
        return F(x // 2)

    return F(x - 1) + F(x + 1)

for x in range(1, 11):
    print(x, F(x))
````

## Section XV. 遞迴寫法檢查清單

寫遞迴時，可以用這個順序檢查：

### 1. 有沒有 base case？

例如：

```python
if n == 1:
    return 1
```

### 2. recursive case 有沒有讓問題變小？

例如：

```python
factorial(n - 1)
```

### 3. 有沒有把遞迴結果接起來？

例如：

```python
return n * factorial(n - 1)
```

### 4. 輸入會不會有不合法情況？

例如：

```python
if n < 0:
    return "不接受負數"
```
