```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
def C(n, m):
    if m == 0 or m == n:
        return 1
    return C(n - 1, m - 1) + C(n - 1, m)

print(C(5, 2))
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

def C(n, m):
    numerator = factorial(n)
    denominator = factorial(m) * factorial(n - m)
    return numerator // denominator

print(C(5, 2))
```

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

print(fib(5))
```

若 x = 1，則 F(x) = 1
若 x 為偶數，則 F(x) = F(x / 2)
其餘狀況，F(x) = F(x - 1) + F(x + 1)
