## Section I. 基礎遞迴：終止條件與問題變小

這一部分先練習最基本的遞迴結構：先判斷什麼時候停止，再呼叫縮小版的自己。


### Q1. 1 到 n 的總和

請完成 `q1_total(n)`，回傳 `1 + 2 + 3 + ... + n`。

提示：`total(n) = n + total(n - 1)`，`total(1) = 1`。


```python
def q1_total(n):
    if n == 1:
        return 1
    return n + q1_total(n - 1)
```

```python
_check_q("q1_total")
```

### Q2. 階乘

請完成 `q2_factorial(n)`，回傳 `n!`。

提示：`0! = 1`，`n! = n * (n - 1)!`。


```python
def q2_factorial(n):
    if n == 0:
        return 1
    return n * q2_factorial(n - 1)
```

```python
_check_q("q2_factorial")
```

### Q3. x 的 n 次方

請完成 `q3_power(x, n)`，回傳 `x` 的 `n` 次方。

提示：`power(x, 0) = 1`。


```python
def q3_power(x, n):
    if n == 0:
        return 1
    return x * q3_power(x, n - 1)
```

```python
_check_q("q3_power")
```

```python
def q4_count_down(n):
    if n == 1:
        return [1]
    return [n] + q4_count_down(n - 1)
```

## Section II. 字串與串列遞迴

這一部分練習把字串或串列切小，例如 `s[1:]`、`nums[1:]`。


### Q9. 反轉字串

請完成 `q9_reverse_string(s)`，回傳反轉後的字串。

例如：`q9_reverse_string("abc")` 應回傳 `"cba"`。


```python
def q9_reverse_string(s):
    if len(s) <= 1:
        return s
    return q9_reverse_string(s[1:]) + s[0]
```

```python
_check_q("q9_reverse_string")
```

### Q10. 判斷回文

請完成 `q10_is_palindrome(s)`，判斷字串是否為回文。

回文是正著讀、反著讀都一樣的字串，例如 `level`。請回傳 `True` 或 `False`。


```python
def q10_is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return q10_is_palindrome(s[1:-1])
```

```python
_check_q("q10_is_palindrome")
```

### Q11. 費氏數列

請完成 `q11_fib(n)`，回傳第 `n` 個費氏數。

定義：`F(0)=0`，`F(1)=1`，`F(n)=F(n-1)+F(n-2)`。


```python
def q11_fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return q11_fib(n - 1) + q11_fib(n - 2)
```

```python
_check_q("q11_fib")
```

### Q12. 串列加總

請完成 `q12_list_sum(nums)`，回傳串列中所有數字的總和。

請用遞迴練習，不要直接使用 `sum(nums)`。


```python
def q12_list_sum(nums):
    if len(nums) == 0:
        return 0
    return nums[0] + q12_list_sum(nums[1:])
```

```python
_check_q("q12_list_sum")
```

### Q13. 找串列最大值

請完成 `q13_max_list(nums)`，回傳串列中的最大值。

假設 `nums` 不會是空串列。


```python
def q13_max_list(nums):
    if len(nums) == 1:
        return nums[0]
    rest_max = q13_max_list(nums[1:])
    if nums[0] > rest_max:
        return nums[0]
    return rest_max
```

```python
_check_q("q13_max_list")
```

### Q14. 計算字元出現次數

請完成 `q14_count_char(s, target)`，回傳 `target` 在字串 `s` 中出現幾次。

例如：`q14_count_char("banana", "a")` 應回傳 `3`。


```python
def q14_count_char(s, target):
    if len(s) == 0:
        return 0
    count = 1 if s[0] == target else 0
    return count + q14_count_char(s[1:], target)
```

## Section III. 進階遞迴與 Online Judge 題型

這一部分練習更常見的競賽題型：最大公因數、快速次方、二進位、組合數與 ZeroJudge 題型。


### Q15. 最大公因數 GCD

請完成 `q15_gcd(a, b)`，使用輾轉相除法回傳 `a` 和 `b` 的最大公因數。

提示：`gcd(a, b) = gcd(b, a % b)`，當 `b == 0` 時停止。


```python
def q15_gcd(a, b):
    if b == 0:
        return a
    return q15_gcd(b, a % b)
```

```python
_check_q("q15_gcd")
```

### Q16. 快速次方

請完成 `q16_fast_power(x, n)`，用快速次方回傳 `x` 的 `n` 次方。

提示：先算 `half = fast_power(x, n // 2)`，再依照 `n` 是奇數或偶數處理。


```python
def q16_fast_power(x, n):
    if n == 0:
        return 1

    half = q16_fast_power(x, n // 2)

    if n % 2 == 0:
        return half * half
    return half * half * x
```

```python
_check_q("q16_fast_power")
```

### Q17. 十進位轉二進位字串

請完成 `q17_binary(n)`，回傳整數 `n` 的二進位字串。

例如：`q17_binary(10)` 應回傳 `"1010"`。


```python
def q17_binary(n):
    if n < 2:
        return str(n)
    return q17_binary(n // 2) + str(n % 2)
```

### Q18. 二項式係數 C(n, m)

請完成 `q18_C(n, m)`，回傳二項式係數 `C(n, m)`。

提示：`C(n,m)=C(n-1,m-1)+C(n-1,m)`，且 `C(n,0)=1`，`C(n,n)=1`。


```python
def q18_C(n, m):
    if m == 0 or m == n:
        return 1
    return q18_C(n - 1, m - 1) + q18_C(n - 1, m)
```

### Q19. ZeroJudge e357 遞迴函數

請完成 `q19_zerojudge_e357(x)`。

定義：若 `x == 1`，回傳 `1`；若 `x` 是偶數，回傳 `F(x // 2)`；其他狀況回傳 `F(x - 1) + F(x + 1)`。


```python
def q19_zerojudge_e357(x):
    if x == 1:
        return 1
    if x % 2 == 0:
        return q19_zerojudge_e357(x // 2)
    return q19_zerojudge_e357(x - 1) + q19_zerojudge_e357(x + 1)
```

```python
def q20_parentheses(n):
    if n == 0:
        return ""
    return "(" + q20_parentheses(n - 2) + ")"
```
